### Calculate Price Momentum (3-5 Year)

```python
def calculate_price_momentum(df):
    """
    Calculate 3-year and 5-year price momentum for each stock.
    Momentum is positive if current price > price 3-5 years ago.
    """
    # Sort by instrument and year
    df_sorted = df.sort_values(['Instrument', 'Year']).copy()
    
    # Calculate momentum by comparing prices
    momentum_data = []
    
    for instrument in df_sorted['Instrument'].unique():
        stock_data = df_sorted[df_sorted['Instrument'] == instrument].copy()
        stock_data = stock_data.sort_values('Year')
        
        for idx, row in stock_data.iterrows():
            current_year = row['Year']
            current_price = row['Close Price']
            
            # Get price 3 years ago
            price_3y_ago = stock_data[stock_data['Year'] == current_year - 3]['Close Price'].values
            price_3y_ago = price_3y_ago[0] if len(price_3y_ago) > 0 else None
            
            # Get price 5 years ago
            price_5y_ago = stock_data[stock_data['Year'] == current_year - 5]['Close Price'].values
            price_5y_ago = price_5y_ago[0] if len(price_5y_ago) > 0 else None
            
            # Calculate momentum
            momentum_3y = None
            momentum_5y = None
            momentum_positive = False
            
            if price_3y_ago and current_price and not pd.isna(price_3y_ago) and not pd.isna(current_price):
                momentum_3y = ((current_price - price_3y_ago) / price_3y_ago) * 100
            
            if price_5y_ago and current_price and not pd.isna(price_5y_ago) and not pd.isna(current_price):
                momentum_5y = ((current_price - price_5y_ago) / price_5y_ago) * 100
            
            # Momentum is positive if either 3Y or 5Y is positive
            if (momentum_3y and momentum_3y > 0) or (momentum_5y and momentum_5y > 0):
                momentum_positive = True
            
            momentum_data.append({
                'Instrument': instrument,
                'Year': current_year,
                'Price_3Y_Momentum_%': momentum_3y,
                'Price_5Y_Momentum_%': momentum_5y,
                'Momentum_Positive': momentum_positive
            })
    
    momentum_df = pd.DataFrame(momentum_data)
    return momentum_df

# Calculate momentum
print("Calculating price momentum...")
momentum_df = calculate_price_momentum(df)

# Merge momentum data back to main dataframe
df = df.merge(momentum_df, on=['Instrument', 'Year'], how='left')

print("\nSample momentum data:")
print(df[df['Momentum_Positive'] == True][['Instrument', 'Year', 'Close Price', 
                                            'Price_3Y_Momentum_%', 'Price_5Y_Momentum_%', 
                                            'Momentum_Positive']].head(10))
```

### Screening Function with All Criteria

```python
def screen_stocks_by_year(df, year):
    """
    Screen stocks based on the criteria for a specific year:
    - Market Cap > 20B
    - D/E < 2
    - ROE: Above Industry Average (6% or 10% threshold)
    - EPS: Positive (+)
    - PE < 25
    - Price momentum 3-5 years: Positive (+)
    - Dividend Yield > 3%
    """
    # Filter for specific year
    yearly_data = df[df['Year'] == year].copy()
    
    # Get industry average ROE for the year
    industry_roe = df[df['Year'] == year]['Return On Equity - Actual'].mean()
    
    # Apply screening criteria
    screened = yearly_data[
        (yearly_data['Company Market Cap'] > 20000) &  # Market cap > 20B (in millions)
        (yearly_data['Total Debt to Common Equity'] < 2) &  # D/E < 2
        (yearly_data['Return On Equity - Actual'] > max(industry_roe, 6)) &  # ROE above industry or 6%
        (yearly_data['Earnings Per Share - Actual'] > 0) &  # EPS positive
        (yearly_data['P/E (Daily Time Series Ratio)'] < 25) &  # PE < 25
        (yearly_data['Momentum_Positive'] == True) &  # Price momentum positive
        (yearly_data['Dividend Per Share Yield %'] > 3)  # Dividend Yield > 3%
    ].copy()
    
    # Select relevant columns
    result = screened[[
        'Instrument', 'Company Name', 'Year', 
        'Company Market Cap', 'Total Debt to Common Equity',
        'Return On Equity - Actual', 'Earnings Per Share - Actual',
        'P/E (Daily Time Series Ratio)', 'Price to Book Value per Share',
        'Dividend Per Share Yield %', 'Close Price',
        'Price_3Y_Momentum_%', 'Price_5Y_Momentum_%'
    ]].sort_values('Company Market Cap', ascending=False)
    
    print(f"\n{'='*80}")
    print(f"Year {int(year)} Screening Results")
    print(f"Industry Average ROE: {industry_roe:.2f}%")
    print(f"Stocks passing criteria: {len(result)}")
    print(f"{'='*80}")
    
    return result

# Screen stocks for each year
years = sorted(df['Year'].dropna().unique(), reverse=True)
screening_results = {}

for year in years:
    screening_results[year] = screen_stocks_by_year(df, year)
    if len(screening_results[year]) > 0:
        print(screening_results[year].to_string())
    else:
        print("No stocks passed the screening criteria for this year.")
```

### Export Results to Excel

```python
# Export screening results to Excel with separate sheets for each year
with pd.ExcelWriter('stock_screening_results_by_year_old_assumption.xlsx', engine='openpyxl') as writer:
    for year in sorted(screening_results.keys(), reverse=True):
        if len(screening_results[year]) > 0:
            screening_results[year].to_excel(writer, sheet_name=f'Year_{int(year)}', index=False)
    
    # Also create a summary sheet
    summary_data = []
    for year in sorted(screening_results.keys(), reverse=True):
        summary_data.append({
            'Year': int(year),
            'Stocks Passing Criteria': len(screening_results[year])
        })
    summary_df = pd.DataFrame(summary_data)
    summary_df.to_excel(writer, sheet_name='Summary', index=False)

print("\nResults exported to 'stock_screening_results_by_year.xlsx'")
```

## Screening with Relaxed Criteria

```python
def screen_stocks_relaxed(df, year, 
                          min_market_cap=10000,  # Relaxed from 20B to 10B
                          max_de=2,              # Keep strict
                          min_roe=5,             # Relaxed from 6 to 5
                          min_eps=0,             # Keep strict (positive)
                          max_pe=30,             # Relaxed from 25 to 30
                          require_momentum=False, # Relaxed - momentum optional
                          min_div_yield=2.5):    # Relaxed from 3% to 2.5%
    """
    Screen stocks with RELAXED criteria:
    - Market Cap > 10B (relaxed from 20B)
    - D/E < 2 (strict)
    - ROE: Above max(Industry Avg, 5%) (relaxed from 6%)
    - EPS: Positive (strict)
    - PE < 30 (relaxed from 25)
    - Price momentum: OPTIONAL (relaxed)
    - Dividend Yield > 2.5% (relaxed from 3%)
    """
    # Filter for specific year
    yearly_data = df[df['Year'] == year].copy()
    
    # Get industry average ROE for the year
    industry_roe = df[df['Year'] == year]['Return On Equity - Actual'].mean()
    
    # Build screening criteria
    criteria = (
        (yearly_data['Company Market Cap'] > min_market_cap) &
        (yearly_data['Total Debt to Common Equity'] < max_de) &
        (yearly_data['Return On Equity - Actual'] > max(industry_roe, min_roe)) &
        (yearly_data['Earnings Per Share - Actual'] > min_eps) &
        (yearly_data['P/E (Daily Time Series Ratio)'] < max_pe) &
        (yearly_data['Dividend Per Share Yield %'] > min_div_yield)
    )
    
    # Add momentum filter only if required
    if require_momentum:
        criteria = criteria & (yearly_data['Momentum_Positive'] == True)
    
    screened = yearly_data[criteria].copy()
    
    # Select relevant columns
    result = screened[[
        'Instrument', 'Company Name', 'Year', 
        'Company Market Cap', 'Total Debt to Common Equity',
        'Return On Equity - Actual', 'Earnings Per Share - Actual',
        'P/E (Daily Time Series Ratio)', 'Price to Book Value per Share',
        'Dividend Per Share Yield %', 'Close Price',
        'Price_3Y_Momentum_%', 'Price_5Y_Momentum_%', 'Momentum_Positive'
    ]].sort_values('Company Market Cap', ascending=False)
    
    print(f"\n{'='*80}")
    print(f"Year {int(year)} - RELAXED Screening Results")
    print(f"Industry Average ROE: {industry_roe:.2f}%")
    print(f"Criteria: MktCap>{min_market_cap/1000}B, D/E<{max_de}, ROE>{min_roe}%, PE<{max_pe}, DivYld>{min_div_yield}%")
    print(f"Momentum Required: {require_momentum}")
    print(f"Stocks passing criteria: {len(result)}")
    print(f"{'='*80}")
    
    return result

# Screen with relaxed criteria
years = sorted(df['Year'].dropna().unique(), reverse=True)
screening_results_relaxed = {}

for year in years:
    screening_results_relaxed[year] = screen_stocks_relaxed(df, year)
    if len(screening_results_relaxed[year]) > 0:
        print(screening_results_relaxed[year].to_string())
    else:
        print("No stocks passed the relaxed screening criteria for this year.")
```
